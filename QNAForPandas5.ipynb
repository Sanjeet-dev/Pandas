# How do I create a dummy variables in pandas?

```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
pd.get_dummies(train,columns=['Sex','Embarked'],drop_first=True)
```

```python
train['Sex_male'] = train.Sex.map({'female':0,'male':1})
```

```python
train.head()
```

```python
train.Embarked.value_counts()
```

```python
embarked_dummies = pd.get_dummies(train.Embarked,prefix='Embarked').astype(float).iloc[:,1:]
```

```python
train = pd.concat([train,embarked_dummies],axis=1)
train.head()
```

```python
pd.get_dummies(train.Sex,prefix='Sex').iloc[:,1:]
```

# How do I find and remove duplicate rows in pandas?

```python
import pandas as pd
```

```python
user_cols =['user_id', 'age', 'gender', 'occupation', 'zip_code']
users =pd.read_csv('http://bit.ly/movieusers', sep='|', header=None, names=user_cols, index_col='user_id')
```

```python
users.head()
```

```python
users.shape
```

```python
users.dtypes
```

```python
users.zip_code.duplicated().sum()
```

```python
# counts only if whole row is identical
users.duplicated().sum() 
```

```python
users.loc[users.duplicated(keep='first'),:]
```

```python
# to drop duplicate one
users.drop_duplicates(keep='first').shape
```

```python
users.duplicated(subset=['age','zip_code']).sum()
```

```python
users.drop_duplicates(subset=['age','zip_code']).shape
```

# How do I fix a settingWithCopy warning?

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.content_rating.isnull().sum()
```

```python
movies[movies.content_rating.isnull()]
```

```python
movies.content_rating.value_counts()
```

```python
import numpy as np
```

```python
movies[movies.content_rating=='NOT RATED'].content_rating = np.nan
```

```python
movies.content_rating.isnull().sum()
```

```python
movies.loc[movies.content_rating=='NOT RATED', 'content_rating'] = np.nan
```

## Another example

```python
top_movies = movies.loc[movies.star_rating >= 9,:]
```

```python
top_movies
```

```python
top_movies.loc[0,'duration'] = 150
```

```python
top_movies
```

```python
top_movies = movies.loc[movies.star_rating >= 9,:].copy()
```

```python
top_movies.loc[0,'duration'] = 150
```

```python
top_movies
```
