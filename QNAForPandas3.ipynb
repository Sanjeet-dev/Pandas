```python
# How do I select multiple rows and columns from a pandas Data Frame?
# loc,iloc
# loc is for selecting things by label like row by index and column by column name
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.loc[0,:]
```

```python
ufo.loc[[0,1,2],:] # or ufo.loc[0:2,:]
```

```python
ufo.loc[:,['City','State']]
```

```python
ufo.loc[0:2,'City':'State'] # or ufo.head(3).drop('Time',axis=1)
```

```python
ufo.head(3).drop('Time',axis=1)
```

```python
ufo[ufo.City=='Oakland'] # or ufo.loc[ufo.City=='Oakland',:]
```

```python
ufo.loc[ufo.City=='Oakland','State']
```

```python
# iloc
# it is for filtering rows and selecting columns by interger position:
```

```python
ufo.iloc[:,[0,3]]
```

```python
ufo.iloc[:,0:3]
```

```python
ufo.iloc[0:3,:]
```

```python
# Don't use this 
ufo[['City','State']] # instead use this ufo.loc[:,['City','State']]
```

```python
ufo[0:2] #instead use this ufo.iloc[0:2,:]
```
